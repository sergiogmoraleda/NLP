PART OF SPEECH

 Part of speech en NLP es una técnica que consiste en etiquetar las palabras de una oración según su categoría gramatical (sustantivo, verbo, adjetivo, etc).

 

```python
import spacy

nlp_en = spacy.load('en_core_web_sm')
nlp_sp = spacy.load('es_core_news_sm')

doc_en = nlp_en(u"The early bird catches the worm.")
doc_sp = nlp_sp(u"A quien madruga, Dios le ayuda.")

def printTokens(doc):
    for token in doc:
        print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop)
        
printTokens(doc_en)
print("\n")
printTokens(doc_sp)
```

```python
from spacy import displacy

doc = nlp_en(u"The quick brown fox jumped over the lazy dog.")
doc_sp = nlp_sp(u"El rápido zorro marrón saltó sobre el perezoso perro.")
```

```python
displacy.render(doc_sp, style="dep", jupyter= True, options={'distance': 120})
```

```python
displacy.render(doc, style="dep", jupyter= True, options={'distance': 120})
```
